```python
# plotting box plots for numerical columns to know about outliers
for i in num:
  sns.boxplot(x=num[i])
  print("boxplot for :"+i)
  plt.show()
```

the box plots are drawn to identify the columns which have outliers
from the graph. the amount_requested,inquiries_last_month, and has_debt columns have outliers hence they need to be checked and treated accordingly

```python
# checking for outlier columns and printing them
for i in num:
    Q1=num[i].quantile(0.25)
    Q3=num[i].quantile(0.75)
    IQR=Q3 - Q1
    lower_bound=Q1 - 1.5 * IQR
    upper_bound=Q3 + 1.5 * IQR

    if (num[(num[i]>upper_bound)]).any(axis=None):
        print(i,":-Outlier Exist")
    elif (num[(num[i]<lower_bound)]).any(axis=None):
        print(i,":-Outlier Exist")
```

```python
# removing rows from columns that are outliers
def oulier(i):
  Q1=fin[i].quantile(0.25)
  Q3=fin[i].quantile(0.75)
  IQR=Q3 - Q1
  lower_bound=Q1 - 1.5 * IQR
  upper_bound=Q3 + 1.5 * IQR
  new=[]
  for ii in num[i]:
      if  (ii>upper_bound):
            new.append(upper_bound)
      elif (ii<lower_bound):
          new.append(lower_bound)
      else:
          new.append(ii)
  fin[i]=new
```

```python
# evaluation measures
def print_scores(alg, y_true, y_pred):
  print(alg)
  alg_names.append(alg)
  acc_score = accuracy_score(y_true, y_pred)
  alg_values.append(acc_score)
  print("accuracy: ",acc_score)
  pre_score = precision_score(y_true, y_pred)
  print("precision: ",pre_score)
  rec_score = recall_score(y_true, y_pred)
  print("recall: ",rec_score)
  f_score = f1_score(y_true, y_pred, average='weighted')
  print("f1_score: ",f_score)
```
